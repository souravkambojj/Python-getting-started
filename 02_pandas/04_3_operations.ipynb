```python
empdata = {'EmpId':[1,2,3,4,5,6,7,8,9],'EmpName':['sam','smith','tom','denis','danny','john','siah','tam','pind'],'salary':[500,700,900,400,1100,210,200,3000,260],'DepId':['e','i','e','i','h','h','e','i','h']}
```

```python
depdata = {'id':['a','e','h','i','j'],'name':['ABC','Engg','HR','IT','JKL']}
```

```python
import pandas as pd

empdf = pd.DataFrame(empdata)
depdf = pd.DataFrame(depdata)
```

```python
print(empdf)
```

```python
print(depdf)
```

```python
depdf.columns
```

# **Rename Columns**

```python
depdf.rename(columns={'id':'DepId','name':'DepName'},inplace=True)
```

# **Merge**

```python
df = pd.merge(empdf,depdf, how='inner', on='DepId')
df
```

# **Sorting**

```python
df.sort_values(by='EmpName')
```

```python
df.sort_values(by='EmpName').reset_index(drop=True) #inplace=True
```

```python
df.sort_values(by='EmpName',ascending=False).reset_index(drop=True)
```

```python
df.sort_values(by=['EmpName','salary'],ascending=[True,False]).reset_index(drop=True)
```
