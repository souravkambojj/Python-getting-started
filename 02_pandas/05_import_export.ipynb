# writing to csv

```python
d1=pd.DataFrame(np.random.rand(4,4),index=['p','q','r','s'],columns=['x','y','z','w'])
d1.to_csv(r'./data/my_csv_sample_file.csv')
pd.read_csv(r'./data/my_csv_sample_file.csv')
```

```python
#index must be false otherwise an index column will be created
d1.to_csv(r'./data/my_csv_sample_file.csv',index=False)
pd.read_csv(r'./data/my_csv_sample_file.csv')
```
