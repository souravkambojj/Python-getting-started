# **concatination**

```python
#make sure dimensions should match along the dir of concatination

import numpy as np
import pandas as pd
from numpy.random import rand as rd

a=rd(4,4)
b=rd(4,4)
c=rd(4,4)
```

```python
d1=pd.DataFrame(a,index=[0,1,2,3],columns=['a','b','c','d'])
d2=pd.DataFrame(b,index=[4,5,6,7],columns=['a','b','c','d'])
d3=pd.DataFrame(c,index=[8,9,10,11],columns=['a','b','c','d'])

print(d1)
print('\n\n',d2)
print('\n\n',d3,'\n\n')
d=pd.concat([d1,d2,d3])
print('after concatination: ')
print(d)
```

```python
#if along the dir of concatination, dimesions are different
#then include ignore_index=True

d1=pd.DataFrame(a,index=[0,1,2,3],columns=['a','b','c','d'])
d2=pd.DataFrame(b,index=[4,5,6,7],columns=['b','a','c','d'])
d3=pd.DataFrame(c,index=[8,9,10,11],columns=['a','b','d','c'])


d=pd.concat([d1,d2,d3],ignore_index=True)
print('after concatination: ')
print(d)
```

# **join**

```python
#join/merge are similar except the key 
#in joins we use  'index' as the key 
#in merge we use a particular column/columns as key

l=pd.DataFrame(np.random.rand(4,2),index=['k0','k1','k2','k4'],columns=['a','b'])
r=pd.DataFrame(np.random.rand(4,2),index=['k0','k2','k3','k4'],columns=['c','d'])

m1=l.join(r,how='left')
m2=l.join(r,how='right')
m3=l.join(r,how='inner')
m4=l.join(r,how='outer')
print(m1)
print(m2)
print(m3)
print(m4)
```

# **Merge**

```python
#merge mtd

left          takes keys of left frame only
right         takes keys of right frame only
inner         takes intersection of keys
outer         takes union of keys
```

```python
k0 k0     k0  k0
k0 k1     k1  k0
k1 k0     k1  k0
k2 k1     k2  k0
```

```python
d1=pd.DataFrame({'key1':['k0','k0','k1','k2'],'key2':['k0','k1','k0','k1'],'a':['a0','a1','a2','a3'],'b':['b0','b1','b2','b3']})
d2=pd.DataFrame({'key1':['k0','k1','k1','k2'],'key2':['k0','k0','k0','k0'],'c':['c0','c1','c2','c3'],'d':['d0','d1','d2','d3']})
```

```python
d1
```

```python
d2
```

```python
pd.merge(d1,d2, how='left', on=['key1','key2'])
```

```python
pd.merge(d1,d2, how='right', on =['key1','key2'])
```

```python
pd.merge(d1,d2, how='inner', on=['key1','key2'])
```

```python
pd.merge(d1,d2, how='outer', on=['key1','key2'])
```
