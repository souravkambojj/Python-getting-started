# Plot Appearance

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,10,1)
y=x**2

fig=plt.figure(figsize=(8,3))
axis=fig.add_axes([0.1,0.2,0.8,0.8])
axis.plot(x,y)
```

```python
x=np.arange(1,10,1)
y=x**2

fig=plt.figure(figsize=(8,3))
axis=fig.add_axes([0.1,0.2,0.8,0.8])
axis.plot(x,y,color='green')
```

```python
x=np.arange(1,10,1)
y=x**2

fig=plt.figure(figsize=(8,3))
axis=fig.add_axes([0.1,0.2,0.8,0.8])
axis.plot(x,y,color='#00FF00')
```

```python
x=np.arange(1,10,1)
y=x**2

fig=plt.figure(figsize=(8,3))
axis=fig.add_axes([0.1,0.2,0.8,0.8])
axis.plot(x,y,color='#00FF00',lw=5)#linewidth or lw : both works
```

```python
x=np.arange(1,10,1)
y=x**2

fig=plt.figure(figsize=(8,3))
axis=fig.add_axes([0.1,0.2,0.8,0.8])
axis.plot(x,y,color='purple',ls=':',marker='*') #ls or linestyle: '--'  '-.'  ':' 'steps'
                                                #marker : it is used to mark each pt on x nd y axis
                                                # '+' 'o' '*' 's' 'l'
```

```python
x=np.arange(1,10,1)
y=x**2

fig=plt.figure(figsize=(8,3))
axis=fig.add_axes([0.1,0.2,0.8,0.8])
axis.plot(x,y,color='purple',ls=':',marker='*',alpha=5) #alpha:it controls how transparent line is
```
