# subplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10,1)
y=x**2
z=x**3
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,z,'y')

plt.tight_layout()  #to provide interspace b/w two subplots

plt.subplot(2,2,3)
plt.plot(y,x,'b')
plt.subplot(2,2,4)
plt.plot(z,x,'r')
```

# object orienting plotting

# mtd1

```python
#add_axes([left,bottom,width,height])
#left,bottom:distance of graph from left and bottom resp.
#width,height:width nd height of box

x=np.arange(1,10,1)
y=x**2

fig=plt.figure()

axis1_big=fig.add_axes([0.1,0.2,0.8,0.9])
axis1_big.set_title('bigger plot')
axis1_big.plot(x,y,'g')


axes2_small=fig.add_axes([0.2,0.5,0.4,0.4])
axes2_small.set_title('smaller plot')
axes2_small.plot(x,y,'r')
```

# plots on same fig

```python
fig=plt.figure(figsize=(4,2))
axis=fig.add_axes([0.1,0.2,2,2])

x=np.arange(1,10,1)
y=x**2
z=x**3
axis.plot(x,y,'g')
axis.plot(x,z,'r')
axis.legend(['square','cube'],loc=10)#to set location of legend there r code 0 to 10
plt.show()
```

```python
x=np.arange(1,10,1)
y=x**2

fig,axis=plt.subplots(1,2)

axis[0].plot(x,y,'g')
axis[0].set_title('square function')

axis[1].set_title('sqrt function')
axis[1].plot(y,x,'r')
```

# figure_size nd DPI

```python
#fig,axis=plt.subplots(n_rows,n_cols,figsize=(n,m),dpi=100) #dpi:dots per inch or pixels per inch
                                            #fig_size(a,b): a & b are width nd height of figure in inches
    
x=np.arange(1,10,1)
y=x**2

fig,axis=plt.subplots(2,1,figsize=(12,6),dpi=100)
axis[0].plot(x,y,'g')
axis[0].set_title('square func')
axis[0].set_xlabel('x-axis')
axis[0].set_ylabel('y-axis')

plt.tight_layout()
axis[1].plot(y,x,'r')
axis[1].set_title('sqrt func')
```

```python
#saving plot as an image
fig.savefig('plot1.png',dp1=100)
```
