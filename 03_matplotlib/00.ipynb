```python
import matplotlib.pyplot as plt
```

# Line plot

```python
x=[2,3,4,7,9]
y=[4,6,9,3,2]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.plot(x,y,marker='*')
plt.show()
```

# Printing data pts only

```python
x=[2,3,4,7,9]
y=[4,6,9,3,2]
plt.plot(x,y,'*')
plt.show()
```

# plotting two or more lines on the same plot

```python
x1=[1,2,3,4,5,6]
y1=[1,8,2,7,3,6]

x2=[2,3,5,7,8,9]
y2=[7,3,8,2,6,1]

plt.plot(x1,y1,':',x2,y2,'-.')
plt.legend(['Line1','Line2'])
plt.show()
```

# Bar Plot

```python
x=[1,2,3,4,5,6]
y=[3,8,5,6,9,2]

plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,color=['red','green','blue'],tick_label=['one','two',3,4.0,'five',6])
plt.show()
```

# Histogram

```python
x=[6,7,3,4,1]
plt.hist(x)
plt.show()
```

# scatter plot

```python
x=[2,4,5,7,8,9]
y=[5,3,7,4,9,8]

plt.scatter(x,y,marker='*',color=['green'])
plt.legend(['stars'])
```
