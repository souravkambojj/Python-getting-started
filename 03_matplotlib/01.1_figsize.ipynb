```
import matplotlib.pyplot as plt
import numpy as np
```

```
plt.figure(figsize=(4,2))
x=np.arange(1,10,1)
y=x**2
z=x**3
plt.plot(x,y,'g')
plt.plot(x,z,'r')
plt.legend(['square','cube'],loc=10)#to set location of legend there r code 0 to 10
plt.show()
```
