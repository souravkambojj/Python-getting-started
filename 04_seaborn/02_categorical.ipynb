```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

# Bar Plot

```python
sns.barplot(df['day'],df['total_bill'])
#here by default estimator is mean()
```

```python
sns.barplot(df['day'],df['total_bill'],estimator=np.sum)
```

```python
sns.barplot(df['day'],df['total_bill'],estimator=np.std)
```

```python
sns.barplot(df['day'],df['total_bill'],estimator=np.max)
```

# Count Plot

```python
sns.countplot(df['day'])#it is morelikely  barplot but here y-axis is count of occurance
```

```python
df['day'].value_counts()
```

**Count Plot is used to plot the count or frequency distribution of each unique observation in categorical data** 

# Box Plot

```python
#this plot is used for distribution of categorical data
           #(categorical, numerical)
sns.boxplot(df['day'],df['total_bill'])
```

```python
sns.boxplot(df['day'],df['total_bill'],hue=df['smoker'])
```

# Violin Plot

```python
sns.violinplot(df['day'],df['total_bill'])
```

```python
sns.violinplot(df['day'],df['total_bill'],hue=df['smoker'])
```

```python
sns.violinplot(df['day'],df['total_bill'],split=True)
```

# Strip Plot

```python
sns.stripplot(df['day'],df['total_bill'],jitter=False) #by default jitter is True
```

```python
sns.stripplot(df['day'],df['total_bill'],jitter=True) #adding jitter noise
```

```python
sns.stripplot(df['day'],df['total_bill'],jitter=True,split=True,hue=df['smoker'])
```

# swarm plot(violin + strip )

```python
sns.swarmplot(df['day'],df['total_bill'])
```

# Factor/Generic Plot

```python
sns.factorplot('day','total_bill',data=df,kind='box')
```

```python
sns.factorplot('day','total_bill',data=df,kind='bar')
```

```python
sns.factorplot('day','total_bill',data=df,kind='violin')
```

```python
sns.factorplot('day','total_bill',data=df,kind='strip')
```

```python
sns.factorplot('day','total_bill',data=df,kind='swarm')
```
