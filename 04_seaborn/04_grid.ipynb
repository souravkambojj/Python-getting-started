```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
df.head()
```

# Pair Plot

```python
sns.pairplot(df,hue='species')
```

# PairGrid

```python
#pairplot is simplified version of pairgrid
#pairplot does a lot of stuff automatically where as in pairgrid u have to do.
```

```python
obj=sns.PairGrid(df)
obj.map(plt.scatter)
```

```python
obj=sns.PairGrid(df)
obj.map_diag(sns.distplot)
obj.map_upper(plt.scatter)
obj.map_lower(sns.kdeplot)
```

# FacetGrid

```python
df=sns.load_dataset('tips')
df.head()
```

```python
obj=sns.FacetGrid(df,col='day',row='smoker',hue='sex')#col will seperate the data of male and female into two separate plots vertically
                                            #row will seperate the data of smoker nd not-smoker into two separates plots horizontally
obj.map(sns.distplot,'total_bill')  #'total_bill': is input column to distribution plot
```
