```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
df.head()
```

# Distribution Plot/Histogram

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'],kde=False,bins=50)
```

# KDE Plot

```python
sns.kdeplot(df['total_bill'])  #kde:kernal density estimation
```

# Rug plot

```python
sns.rugplot(df['total_bill'])
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

# Joint Plot

```python
sns.jointplot(df['total_bill'],df['tip'],kind='reg')
```

```python
sns.jointplot(df['total_bill'],df['tip'],kind='kde')
```

```python
sns.jointplot(df['total_bill'],df['tip'],kind='hex')
```
