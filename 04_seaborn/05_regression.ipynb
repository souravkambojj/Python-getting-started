# Regression plots

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df.head()
```

# LM Plot

```python
sns.lmplot('total_bill','tip',data=df,hue='sex')
```

```python
sns.lmplot('total_bill','tip',data=df,col='sex')
```

```python
sns.lmplot('total_bill','tip',data=df,col='sex',row='time')
```

```python
sns.lmplot('total_bill','tip',data=df,col='sex',row='time',hue='day')
```
